###Exemplo

O sistema de uma empresa categoriza seus funcionários por tempo de empresa.
São 4 classes que variam de acordo com o tempo de experiencia que o funcionário tem de empresa.



*   Treinee: até 2 anos
*   Junior:2 a 4 anos
*   Pleno: 4 a 8 anos
*   Sênior:Acima de 8



```python
exp = float(input('Digite quantos anos de empresa você tem? '))

if exp <= 2:
  print('Treinee')
elif 2 < exp <= 4:
  print('Junior')
elif 4 < exp <= 8:
  print('Pleno')
else :
  print('Sênior')
```

```python
faturamento = float(input('Digite o valor do faturamento em vendas: '))
salario_fixo = float(input('Digite o valor do salário fixo do vendedor: '))

if faturamento <= 30000:
  comissao = faturamento * 0.03
  print(f'Sua comissão é de: {round(comissao)}')
elif faturamento <= 60000:
  comissao = faturamento * 0.05
  print(f'Sua comissão é de: {round(comissao)}')
elif faturamento > 60000:
  comissao = faturamento * 0.07
  print(f'Sua comissão é de: {round(comissao)}')

sal_total = salario_fixo + comissao
print(f'O salário total do vendedor é de: {round(sal_total)}')

if salario_fixo > 15000:
    print('Parabéns ! Você teve um excelente mês !')

```
